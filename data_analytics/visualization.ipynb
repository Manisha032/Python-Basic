```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
y = [21,345,56,12,5,47,1]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.pie(y)
plt.show()
```

```python
figure, ax = plt.subplots(nrows=1, ncols=2, figsize =(15,2))
plt.show()
```

```python
t = np.arange(-3* np.pi, 3 * np.pi, 0.2)
x= np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,6))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```

```python
x= [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56,]
plt.plot(x,y, 'go--',linewidth=2, markersize=10, 
         markerfacecolor='white', markeredgecolor='red',)
plt.show()
```

```python
plt.plot(x,y, 'go--',linewidth=2, markersize=10, 
         markerfacecolor='white',)
plt.title("Some graph showing",
          fontdict= {'fontsize':15,'fontweight':'light','color':'green'})
plt.xlabel("X axis ki values")
plt.ylabel("Y axis ki values", fontdict ={'color':'red'})
plt.text(800,60,"This is serious\ntext here")
plt.annotate("Peak value",xy=(500,65), xytext=(600,50),
    arrowprops=dict(facecolor='black', width=1))
plt.show()
```
